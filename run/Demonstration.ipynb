```python
import os
name = graph_visual_data['name']
nodeNameLocation = os.path.join("datasets",name,"processed","nodeNames.pt")
nodeNames = torch.load(nodeNameLocation)
inv_map = {v: k for k, v in nodeNames.items()}
edge_weights = graph_visual_data['edge_weights']

# Convert tensor to numpy array for easier manipulation
edges_np = edges.numpy()

# Create a new list to store the mapped names
named_edges = []

# Loop through the edges and map the indices to names
for edge in edges_np.T:  # Transpose to iterate over pairs of indices
    named_edges.append((inv_map[edge[0]], inv_map[edge[1]]))

# Optionally convert back to a tensor
# named_edges_tensor = torch.tensor(named_edges)

file = open("Weights.txt", mode = "w")
file.write("Source,Dest,Weight\n")
weight_list = [(named_edges[i][0], named_edges[i][1], edge_weights[i].item()) for i in range(len(edge_weights))]
weight_list = sorted(weight_list, key=lambda x: -abs(x[2]))
for i, weight in enumerate(edge_weights):  
    file.write(weight_list[i][0].ljust(25) + "," + weight_list[i][1].ljust(25) + "," + str(weight_list[i][2]) + "\n")
```
